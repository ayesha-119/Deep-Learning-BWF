# **Supervised Learning:**

**Definition:** Learning to map input data to known targets with labeled examples.

**Examples:** Binary classification, multiclass classification, scalar regression.

**Applications:** Optical character recognition, speech recognition, image classification, language translation.


# **Unsupervised Learning:**

**Definition:** Finding interesting transformations of input data without targets.

**Examples:** Dimensionality reduction, clustering.

**Applications:** Data visualization, data compression, data denoising, understanding correlations in data.


# **Self-Supervised Learning:**

**Definition:** Supervised learning without human-annotated labels.

**Examples:** Autoencoders, predicting next frames in a video, predicting next words in a text.

**Applications:** Learning useful representations, pretraining models.


# **Reinforcement Learning:**

**Definition:** Learning to choose actions to maximize rewards in an environment.

**Examples:** Training a neural network to play games.

**Applications:** Self-driving cars, robotics, resource management, education.
Note: The boundaries between these branches can be blurry, and there can be overlap or variations in specific applications.
