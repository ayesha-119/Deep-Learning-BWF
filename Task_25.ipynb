**What is Overfitting?**


Overfitting occurs when a model becomes too complex and starts to memorize the training data instead of learning generalizable patterns.
The model performs well on the training data but fails to generalize to new, unseen data.

**Causes of Overfitting:**

Insufficient training data: When the amount of training data is limited, the model may try to fit noise or outliers, resulting in overfitting.
Complex models: Models with a large number of parameters or high complexity can easily overfit the training data.

**Consequences of Overfitting:**

Poor generalization: Overfitted models fail to generalize to new data, leading to reduced performance and unreliable predictions.
High variance: Overfitted models are highly sensitive to small changes in the training data, causing significant fluctuations in performance.



**What is Underfitting?**

Underfitting occurs when a model is too simple to capture the underlying patterns in the data.
The model performs poorly on both the training data and new data due to its limited capacity.

**Causes of Underfitting:**

Insufficient model capacity: If the model is too simple or lacks the necessary complexity, it may struggle to capture the underlying patterns in the data.

**Consequences of Underfitting:**

Inability to learn complex relationships: Underfitted models cannot capture the nuances and complexities present in the data, resulting in suboptimal performance.

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize pixel values between 0 and 1
x_train = x_train / 255.0
x_test = x_test / 255.0

# Example 1: Overfitting
# Define a deep learning model with more complexity
overfit_model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Compile the overfitting model
overfit_model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# Train the overfitting model
history_overfit = overfit_model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))

# Evaluate the overfitting model
test_loss_overfit, test_accuracy_overfit = overfit_model.evaluate(x_test, y_test)

# Plot the training and test accuracy and loss curves for the overfitting model
plt.plot(history_overfit.history['accuracy'])
plt.plot(history_overfit.history['val_accuracy'])
plt.title('Overfitting Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(history_overfit.history['loss'])
plt.plot(history_overfit.history['val_loss'])
plt.title('Overfitting Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()

# Example 2: Underfitting
# Define a simple model with limited capacity
underfit_model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(10, activation='softmax')
])

# Compile the underfitting model
underfit_model.compile(optimizer='adam',
                       loss='sparse_categorical_crossentropy',
                       metrics=['accuracy'])

# Train the underfitting model
history_underfit = underfit_model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))

# Evaluate the underfitting model
test_loss_underfit, test_accuracy_underfit = underfit_model.evaluate(x_test, y_test)

# Plot the training and test accuracy and loss curves for the underfitting model
plt.plot(history_underfit.history['accuracy'])
plt.plot(history_underfit.history['val_accuracy'])
plt.title('Underfitting Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(history_underfit.history['loss'])
plt.plot(history_underfit.history['val_loss'])
plt.title('Underfitting Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```

**Adding weight regularization**

```python
from tensorflow.keras import regularizers
# Example 1: Overfitting with L2 regularization
overfit_model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    keras.layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    keras.layers.Dense(10, activation='softmax')
])

# Compile the overfitting model
overfit_model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

# Train the overfitting model
history_overfit = overfit_model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))

# Evaluate the overfitting model
test_loss_overfit, test_accuracy_overfit = overfit_model.evaluate(x_test, y_test)

# Plot the training and test accuracy and loss curves for the overfitting model
plt.plot(history_overfit.history['accuracy'])
plt.plot(history_overfit.history['val_accuracy'])
plt.title('Overfitting Model Accuracy after Regularization')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(history_overfit.history['loss'])
plt.plot(history_overfit.history['val_loss'])
plt.title('Overfitting Model Loss after Regularization')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```

# **Adding dropout**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import matplotlib.pyplot as plt

# Set parameters for the IMDB dataset
max_words = 10000
max_length = 200

# Load the IMDB dataset
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_words)

# Preprocess the data by padding the sequences
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_length)
x_test = keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_length)

# Train the model without dropout
model_no_dropout = Sequential([
    Dense(128, activation='relu', input_shape=(max_length,)),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model without dropout
model_no_dropout.compile(optimizer='adam',
                         loss='binary_crossentropy',
                         metrics=['accuracy'])

# Train the model without dropout
history_no_dropout = model_no_dropout.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))

# Train the model with dropout
model_with_dropout = Sequential([
    Dense(128, activation='relu', input_shape=(max_length,)),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

# Compile the model with dropout
model_with_dropout.compile(optimizer='adam',
                           loss='binary_crossentropy',
                           metrics=['accuracy'])

# Train the model with dropout
history_with_dropout = model_with_dropout.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))

# Plot the validation loss curves for comparison
plt.plot(history_no_dropout.history['val_loss'], label='No Dropout')
plt.plot(history_with_dropout.history['val_loss'], label='With Dropout')
plt.title('Validation Loss Comparison')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
