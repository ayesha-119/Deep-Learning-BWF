**Element-wise addition**

```python
import numpy as np

def elementwise_addition(x, y):
    """
    Performs element-wise addition of two tensors x and y.
    """
    assert x.shape == y.shape, "Shapes of tensors must match."
    
    result = np.zeros_like(x)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            result[i, j] = x[i, j] + y[i, j]
    
    return result
```

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])
result = elementwise_addition(x, y)
print(result)
```

**Element-wise multiplication and subtraction**

```python
import numpy as np

# Define two sample 2D tensors
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([[7, 8, 9], [10, 11, 12]])

# Element-wise multiplication using for loop
result_mult = np.zeros_like(x)
for i in range(x.shape[0]):
    for j in range(x.shape[1]):
        result_mult[i, j] = x[i, j] * y[i, j]

# Element-wise subtraction using for loop
result_sub = np.zeros_like(x)
for i in range(x.shape[0]):
    for j in range(x.shape[1]):
        result_sub[i, j] = x[i, j] - y[i, j]

# Print the original and the result tensors for multiplication and subtraction
print("Original tensors:\n", x, "\n", y)
print("\nResult tensor (multiplication):\n", result_mult)
print("\nResult tensor (subtraction):\n", result_sub)
```

# **Broadcasting**

```python
def naive_add_matrix_and_vector(x, y):
    assert len(x.shape) == 2
    assert len(y.shape) == 1
    assert x.shape[1] == y.shape[0]
    
    x = x.copy()
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            x[i, j] += y[j]
            
    return x
```

```python
import numpy as np

# create a matrix A of shape (2, 3)
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# create a vector b of shape (3,)
b = np.array([10, 20, 30])

# add b to each row of A using naive_add_matrix_and_vector
result = naive_add_matrix_and_vector(A, b)

print(result)
```

# **Dot Product:**

```python
def naive_dot(x, y):
    assert len(x.shape) == 1
    assert len(y.shape) == 1
    assert x.shape[0] == y.shape[0]
    
    result = 0
    for i in range(x.shape[0]):
        result += x[i] * y[i]
    
    return result
```

```python
import numpy as np

# define two one-dimensional arrays of length 3
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

# compute the dot product using the naive_dot function
dot_product = naive_dot(x, y)

# print the result
print(dot_product)
```
